```python
!python - <<'PY'
import numpy as np, onnxruntime as ort
p="/kaggle/working/EXPORT/runtime_bundle/pill_cls.onnx"
sess=ort.InferenceSession(p, providers=["CPUExecutionProvider"])
print("inputs:", [(i.name,i.shape,i.type) for i in sess.get_inputs()])
print("outputs:", [(o.name,o.shape,o.type) for o in sess.get_outputs()])
name=sess.get_inputs()[0].name
# assume NCHW float32
x=np.zeros((1,3,448,448), dtype=np.float32)
y=sess.run(None, {name:x})
print("ok, output shape:", y[0].shape)
```
